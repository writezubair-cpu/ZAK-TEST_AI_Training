Learn about conditional statements (if-else) and loops (for, while) to control the execution of your code.

```python
#video 4
from IPython.display import HTML

HTML("""
<iframe width="560" height="315" src="https://www.youtube.com/embed/qDgkeETAiNQ?si=0tHrsuKEb_qovxGj" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" referrerpolicy="strict-origin-when-cross-origin" allowfullscreen></iframe>
""")
```

```python
# 1. Demonstrate conditional statements (if-elif-else)

# Example with numerical values
temperature = 20

if temperature > 30:
  print("It's a hot day!")
elif temperature > 20:
  print("It's a pleasant day.")
else:
  print("It's a bit chilly.")

# Example with string values
color = "blue"

if color == "red":
  print("The color is red.")
elif color == "blue":
  print("The color is blue.")
else:
  print("The color is neither red nor blue.")


# 2. Create a for loop
fruits = ["apple", "banana", "cherry"]

print("\nIterating through the fruits list:")
for fruit in fruits:
  print(fruit)


# 3. Write a while loop
count = 0

print("\nCounting with a while loop:")
while count < 5:
  print(count)
  count += 1 # Increment count to eventually break the loop
```

```python
# 1. Define a simple function named greet
def greet(name):
  """Greets a person with a given name."""
  print(f"Hello ENEC, {name}!")

# 2. Call the greet function with your name
greet("Mohammed Alkhatri 1111111111111")

# 3. Define a function named add_numbers
def add_numbers(num1, num2):
  """Adds two numbers and returns their sum."""
  return num1 + num2

# 4. Call the add_numbers function with two numbers and print the returned result
result_sum = add_numbers(100, 100)
print(f"The sum of 100 and 100 is: {result_sum}")

# 5. Define a function named calculate_area with default arguments
def calculate_area(length=1, width=1):
  """Calculates the area of a rectangle with optional default values."""
  return length * width

# 6. Call the calculate_area function without arguments and print the result
area_default = calculate_area()
print(f"Area with default values: {area_default}")

# 7. Call the calculate_area function with specific length and width values and print the result
area_specific = calculate_area(length=10, width=5)
print(f"Area with specific values (10x5): {area_specific}")
```
