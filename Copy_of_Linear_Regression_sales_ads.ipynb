````python


# ------------------------------------------------------------------------------
# STEP 2: Define the Data (X and Y)
# We define the inputs (X, the budget) and the output (Y, the sales) [2].
# ------------------------------------------------------------------------------

# X (Inputs/Features): TV and Radio budget
X = df[['TV_Budget', 'Radio_Budget']]

# Y (Output/Target): Sales
Y = df['Sales']

print(f"\n# Inputs (X) defined with shape: {X.shape}")
print(f"# Output (Y) defined with shape: {Y.shape}")


# ------------------------------------------------------------------------------
# STEP 3: Create and Fit the Model
# We use .fit to tell the model to 'go learn the pattern from this data' [3].
# ------------------------------------------------------------------------------

model = LinearRegression()
model.fit(X, Y)

print("\n# Model successfully created and fitted (learned the pattern).")

# ------------------------------------------------------------------------------
# STEP 4: Print Results (Continued)
# ------------------------------------------------------------------------------

# --- A. Intercept (The Baseline) ---
intercept = model.intercept_
print(f"\n# -------------------------------------------------------------")
print(f"# Learned Intercept (Baseline, B): {intercept:.2f}")

# --- B. Coefficients (The Weights / ROI) ---
coefficients = model.coef_

# Extract coefficients by index:
# FIX: Use index 0 for the first input (TV_Budget)
tv_coef = coefficients[0]
# FIX: Use index 1 for the second input (Radio_Budget). Index 5 was Out of Bounds.
radio_coef = coefficients[1]

print(f"# Learned Coefficients (ROI per input):")
print(f"#   TV_Budget: {tv_coef:.2f}")
print(f"#   Radio_Budget: {radio_coef:.2f}")
print(f"# -------------------------------------------------------------")

# ------------------------------------------------------------------------------
# INTERPRETATION SUMMARY
# ------------------------------------------------------------------------------

print("\n# Final Model Interpretation (The Line of Best Fit Equation):")
# Use the extracted individual coefficients for the final equation printout
print(f"# Estimated Sales = ({tv_coef:.2f} * TV_Budget) + ({radio_coef:.2f} * Radio_Budget) + {intercept:.2f}")
````

````python
import matplotlib.pyplot as plt

# Generate predictions using the fitted model
predictions = model.predict(X)

# Create a scatter plot of Actual Sales vs. Predicted Sales
plt.figure(figsize=(10, 6))
plt.scatter(Y, predictions, alpha=0.7)

# Add a line representing a perfect prediction (y=x)
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--', lw=2, label='Perfect Prediction Line')

plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs. Predicted Sales from Linear Regression Model')
plt.grid(True)
plt.legend()
plt.show()
````

### Clarify the nature of the 'vibration' data

Given that no specific vibration data has been provided, we will proceed by generating a synthetic dataset to demonstrate the analysis workflow. This synthetic data will be a **time-series dataset** simulating acceleration measurements.

Here are the details for the synthetic data generation:

1.  **Type of Measurement**: The data will represent **acceleration**.
2.  **Units**: Measurements will be in **g (gravitational force)**.
3.  **Sampling Rate**: A simulated sampling rate of **1000 Hz (1000 samples per second)** will be used.
4.  **Duration**: The total duration of the simulated vibration recording will be **10 seconds**.
5.  **Characteristics**: The vibration data will include:
    *   A **dominant frequency** (e.g., 50 Hz) to simulate a rotating component.
    *   **Harmonics** of the dominant frequency (e.g., 100 Hz, 150 Hz) to represent non-linearities or other machine components.
    *   **Random noise** to simulate real-world measurement imperfections.
    *   A **transient event** (a sudden spike) to represent an anomaly or impact.

This will allow us to proceed with an example that reflects typical vibration analysis scenarios.

**Reasoning**:
I will generate synthetic vibration data based on the characteristics outlined previously, including a dominant frequency, harmonics, noise, and a transient event, and store it in a Pandas DataFrame.



````python
import numpy as np
import pandas as pd

# Parameters for synthetic data generation
SAMPLING_RATE = 1000  # Hz
DURATION = 10         # seconds
NUM_SAMPLES = SAMPLING_RATE * DURATION
time = np.linspace(0, DURATION, NUM_SAMPLES, endpoint=False)

# 1. Dominant Frequency Component (e.g., from a rotating component)
dominant_freq = 50  # Hz
amplitude_dominant = 1.0 # g
signal_dominant = amplitude_dominant * np.sin(2 * np.pi * dominant_freq * time)

# 2. Harmonics (e.g., 2nd and 3rd harmonics)
harmonic_freq_2 = dominant_freq * 2 # 100 Hz
amplitude_harmonic_2 = 0.4 # g
signal_harmonic_2 = amplitude_harmonic_2 * np.sin(2 * np.pi * harmonic_freq_2 * time)

harmonic_freq_3 = dominant_freq * 3 # 150 Hz
amplitude_harmonic_3 = 0.2 # g
signal_harmonic_3 = amplitude_harmonic_3 * np.sin(2 * np.pi * harmonic_freq_3 * time)

# 3. Random Noise
noise_amplitude = 0.1 # g
random_noise = noise_amplitude * np.random.randn(NUM_SAMPLES)

# 4. Transient Event (a sudden spike)
transient_start_time = 5.0 # seconds
transient_duration = 0.05 # seconds
transient_amplitude = 5.0 # g

transient_start_idx = int(transient_start_time * SAMPLING_RATE)
transient_end_idx = int((transient_start_time + transient_duration) * SAMPLING_RATE)

transient_signal = np.zeros(NUM_SAMPLES)
if transient_end_idx < NUM_SAMPLES:
    transient_signal[transient_start_idx:transient_end_idx] = transient_amplitude * np.hanning(transient_end_idx - transient_start_idx)

# Combine all components to create the final synthetic vibration signal
synthetic_vibration_g = (
    signal_dominant +
    signal_harmonic_2 +
    signal_harmonic_3 +
    random_noise +
    transient_signal
)

# Create a Pandas DataFrame
vibration_df = pd.DataFrame({
    'Time_s': time,
    'Acceleration_g': synthetic_vibration_g
})

print("# Synthetic vibration data generated.")
print(f"# Data points: {len(vibration_df)}")
print("# First 5 rows of the generated data:")
print(vibration_df.head())

# Display basic statistics
print("\n# Descriptive statistics of the generated data:")
print(vibration_df.describe())
````

## Frequency Domain Analysis

### Subtask:
Apply a Fast Fourier Transform (FFT) to the 'Acceleration_g' data to analyze the signal in the frequency domain. Plot the amplitude spectrum to identify dominant frequencies and their harmonics. Ensure the x-axis represents frequency (Hz) and the y-axis represents amplitude or magnitude.

#### Instructions
1. Calculate the Fast Fourier Transform (FFT) of the 'Acceleration_g' signal.
2. Compute the corresponding frequencies for the FFT result.
3. Plot the amplitude spectrum (magnitude of the FFT).
4. Label the x-axis as 'Frequency (Hz)' and the y-axis as 'Amplitude'.
5. Add a title to the plot, such as 'Frequency Spectrum of Synthetic Vibration Data'.

**Reasoning**:
I will calculate the Fast Fourier Transform (FFT) of the 'Acceleration_g' signal, compute the corresponding frequencies, and then plot the amplitude spectrum to visualize the dominant frequencies as requested by the subtask. I will include appropriate labels and title.



````python
import numpy as np
import matplotlib.pyplot as plt

# Get the acceleration data
acceleration_signal = vibration_df['Acceleration_g'].values

# Number of samples and sampling rate are already defined
# NUM_SAMPLES = SAMPLING_RATE * DURATION
# SAMPLING_RATE = 1000 # Hz

# 1. Calculate the Fast Fourier Transform (FFT)
fft_result = np.fft.fft(acceleration_signal)

# 2. Compute the corresponding frequencies
# The frequencies correspond to the FFT result. fftfreq returns the sample frequencies.
freqs = np.fft.fftfreq(NUM_SAMPLES, d=1/SAMPLING_RATE)

# 3. Plot the amplitude spectrum (magnitude of the FFT)
# We only care about the positive frequencies for real-valued signals.
# The first half of the FFT result (up to N/2) contains the unique information.
positive_freq_indices = np.where(freqs >= 0)
freqs_positive = freqs[positive_freq_indices]
amplitude_spectrum = np.abs(fft_result[positive_freq_indices]) * (1 / NUM_SAMPLES) # Normalize by N

plt.figure(figsize=(12, 6))
plt.plot(freqs_positive, amplitude_spectrum, color='red', linewidth=0.8)
plt.title('Frequency Spectrum of Synthetic Vibration Data')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (g)')
plt.grid(True)
plt.xlim(0, SAMPLING_RATE / 2) # Limit x-axis to Nyquist frequency

# Highlight expected dominant frequencies from synthetic data generation
plt.axvline(x=dominant_freq, color='green', linestyle='--', label=f'Dominant Freq: {dominant_freq} Hz')
plt.axvline(x=harmonic_freq_2, color='purple', linestyle='--', label=f'2nd Harmonic: {harmonic_freq_2} Hz')
plt.axvline(x=harmonic_freq_3, color='orange', linestyle='--', label=f'3rd Harmonic: {harmonic_freq_3} Hz')
plt.legend()

plt.show()
````
