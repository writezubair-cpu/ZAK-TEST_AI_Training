```python
# Import the necessary data analysis library (pandas is effective for descriptive statistics)
import pandas as pd

# 1. CREATE THE SAMPLE DATA
# Salary in thousands (K). Most values are around 25–95K,
data = {
    'Salary_K': [
        25, 28, 30, 32, 35, 35, 38, 40, 42, 45,
        48, 50, 52, 55, 60, 65, 70, 80, 95
    ]
}

# data_outlier = {
#     'Salary_K': [
#         25, 28, 30, 32, 35, 35, 38, 40, 42, 45,
#         48, 50, 52, 55, 60, 65, 70, 80, 95, 1000  # 1000 is the outlier
#     ]
# }
df = pd.DataFrame(data)

print("Created Sample Dataset:")
print(df)
print("-" * 50)

# 2. CALCULATE DESCRIPTIVE STATISTICS

# A. Measures of Central Tendency (Describing the center of the data)
print("A. Central Tendency (Center):")
mean_salary = df['Salary_K'].mean()
median_salary = df['Salary_K'].median()
mode_salary = df['Salary_K'].mode()

print(f"Mean (Arithmetic Average): {mean_salary:.2f} K")  # Pulled up by 500
print(f"Median (Middle Value): {median_salary:.2f} K")    # More robust here
print(f"Mode (Most Frequent Value): {list(mode_salary)}")
print("-" * 50)

# B. Measures of Dispersion (Spread)
print("B. Measures of Dispersion (Spread):")

std_dev = df['Salary_K'].std()
variance = df['Salary_K'].var()
data_range = df['Salary_K'].max() - df['Salary_K'].min()

Q1 = df['Salary_K'].quantile(0.25)
Q3 = df['Salary_K'].quantile(0.75)
IQR = Q3 - Q1

print(f"Standard Deviation: {std_dev:.2f} K")
print(f"Variance: {variance:.2f} (Squared K units)")
print(f"Range: {data_range:.2f} K")
print(f"Interquartile Range (IQR): {IQR:.2f} K")
print("-" * 50)

# C. Quick pandas summary
print("\nC. Comprehensive Descriptive Summary (df.describe()):")
print(df['Salary_K'].describe())

```

```python
from IPython.display import HTML

HTML("""
<iframe width="560" height="315" src="https://www.youtube.com/embed/8fF2Dr4mjwQ?si=kxyZxmzPV1rnTzcl" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" referrerpolicy="strict-origin-when-cross-origin" allowfullscreen></iframe>
""")
```

```python
# Step 1: Import the libraries we need
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# Step 2: Set up our "Abu Dhabi salary" assumptions
# ------------------------------
# We are assuming (for teaching purposes):
# - Average monthly salary in Abu Dhabi = 20,000 AED
# - Standard deviation (spread) = 5,000 AED
# - We will simulate 10,000 people

mean_salary = 20000   # center of the distribution (AED)
std_salary = 5000     # spread of the distribution (AED)
sample_size = 10000   # how many salary values to generate

# Optional: This makes the random results repeatable every time you run the code
np.random.seed(42)

# ------------------------------
# Step 3: Simulate salary data
# ------------------------------
# We use a Normal Distribution to generate "fake" Abu Dhabi salaries.
# loc  = mean of the distribution
# scale = standard deviation
# size = number of samples

raw_salaries = np.random.normal(
    loc=mean_salary,
    scale=std_salary,
    size=sample_size
)

# Realistically, salaries can't be negative or extremely tiny,
# so we clip very low values to a minimum salary (e.g., 3,000 AED).
min_salary = 3000
salaries = np.clip(raw_salaries, min_salary, None)

# ------------------------------
# Step 4: Print some basic stats so students see what we created
# ------------------------------
print(f"Number of people in sample: {len(salaries)}")
print(f"Sample mean salary: {np.mean(salaries):.2f} AED")
print(f"Sample standard deviation: {np.std(salaries, ddof=1):.2f} AED")

# ------------------------------
# Step 5: Plot the histogram (this shows the shape of the data)
# ------------------------------
plt.figure(figsize=(10, 6))

# density=True → y-axis shows probability density, not just raw counts
counts, bin_edges, _ = plt.hist(
    salaries,
    bins=40,
    density=True,
    alpha=0.6,
    edgecolor='black',
    label='Simulated salaries (histogram)'
)

# ------------------------------
# Step 6: Draw the smooth Normal (bell) curve on top
# ------------------------------
# We calculate the theoretical Normal PDF with the same mean and std we used.

# Create 200 x-values between the min and max salary for a smooth curve
x = np.linspace(salaries.min(), salaries.max(), 200)

# Normal distribution formula (Probability Density Function)
pdf = (1 / (std_salary * np.sqrt(2 * np.pi))) * np.exp(
    -0.5 * ((x - mean_salary) / std_salary) ** 2
)

plt.plot(x, pdf, linewidth=2, label='Theoretical bell curve')

# ------------------------------
# Step 7: Make the plot look nice and informative
# ------------------------------
plt.title("Simulated Monthly Salaries in Abu Dhabi (AED)")
plt.xlabel("Monthly Salary (AED)")
plt.ylabel("Density (Probability)")
plt.grid(axis='y', alpha=0.4)
plt.legend()

# ------------------------------
# Step 8: Show the final plot
# ------------------------------
plt.show()
```

```python
# --- 2) Create a simple sample dataset + quick 95% CI for the mean ---
import numpy as np
import pandas as pd

np.random.seed(42)

# Skewed population (shows why CLT helps)
population = np.random.lognormal(mean=0.0, sigma=0.9, size=100_000)

# Take one sample
n = 200
sample = np.random.choice(population, size=n, replace=False)

# DataFrame and save
df = pd.DataFrame({"value": sample})
df.to_csv("sample_clt_ci_data.csv", index=False)

# Quick 95% CI (normal approx.)
xbar = df["value"].mean()
s = df["value"].std(ddof=1)
se = s / np.sqrt(n)
ci_low, ci_high = xbar - 1.96 * se, xbar + 1.96 * se

print("First 5 rows:")
print(df.head(), "\n")
print(f"n = {n}")
print(f"mean = {xbar:.4f}")
print(f"std  = {s:.4f}")
print(f"SE   = {se:.4f}")
print(f"95% CI for mean: ({ci_low:.4f}, {ci_high:.4f})")
print("Saved CSV: sample_clt_ci_data.csv")
```

```python
#Virtual 10 Coin Flip Simulation Script (Conceptual Example)
import numpy as np
# The sources suggest using numpy.random for simulations [1].

# 1. Define the parameters
NUM_FLIPS = 10000
# The goal is to simulate 10,000 coin tosses [1].

# 2. Simulate the coin flips
# We use numpy.random.randint to generate 10000 random integers (0 or 1).
# We can define 0 as 'Tails' and 1 as 'Heads'.
results = np.random.randint(0, 2, size=NUM_FLIPS)

# 3. Calculate the outcomes
heads_count = np.sum(results)
tails_count = NUM_FLIPS - heads_count

# 4. Determine the relative frequency
# The simulation aims to see the frequency of heads ~0.5 [1].
frequency_heads = heads_count / NUM_FLIPS
frequency_tails = tails_count / NUM_FLIPS

# 5. Output the results
print(f"Simulation of {NUM_FLIPS} Coin Flips")
print("---------------------------------")
print(f"Total Heads Count: {heads_count}")
print(f"Total Tails Count: {tails_count}")
print(f"Relative Frequency of Heads: {frequency_heads:.4f}")
print(f"Relative Frequency of Tails: {frequency_tails:.4f}")
print("\nThis demonstrates the Law of Large Numbers in action.")
```

# 6 Hypothesis Testing and Statistical Significance

```python
from IPython.display import HTML
HTML("""
<iframe width="560" height="315" src="https://www.youtube.com/embed/lPSEjbXrZRM?si=ljiJRe9FshqwlSbN" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" referrerpolicy="strict-origin-when-cross-origin" allowfullscreen></iframe>
""")

```

```python

# ----------------------------------------------------
# Step 1: Import Necessary Libraries
# We need numpy for creating sample data and scipy.stats
# to perform the T-test (the "workhorse" for comparing two averages).
# ----------------------------------------------------
import numpy as np
from scipy import stats

# Set a seed for reproducibility so the 'random luck' data is consistent
np.random.seed(42)

# ----------------------------------------------------
# Step 2: Create Sample Data (Ad A vs. Ad B Clicks)
# This simulates the data we "hand" to the SciPy library.
# We create data for Ad A (Old Ad) and Ad B (New Ad).
# We assume Ad B has a slightly higher average, reflecting the hunch.
# ----------------------------------------------------

# Ad A (Control/Old Ad): Average daily clicks around 50
ad_a_clicks = np.random.normal(loc=50, scale=10, size=100)

# Ad B (Treatment/New Ad): Average daily clicks around 55 (simulating a positive effect)
ad_b_clicks = np.random.normal(loc=55, scale=10, size=100)

# Print a quick summary to see the difference in averages
print("--- Sample Data Summary ---")
print(f"Ad A Average Clicks: {ad_a_clicks.mean():.2f}")
print(f"Ad B Average Clicks: {ad_b_clicks.mean():.2f}")
print("---------------------------\n")

# ----------------------------------------------------
# Step 3: Run the Hypothesis Test (The single line of code)
# We run the T-test, comparing the two independent samples.
# This is where the Python library does "all the hard math for you".
# ----------------------------------------------------

# The ttest_ind function returns the Test Statistic and the P value.
test_statistic, p_value = stats.ttest_ind(ad_a_clicks, ad_b_clicks)

# ----------------------------------------------------
# Step 4: Interpret the Results (The Big Finale)
# We compare the P value to the significance level (alpha = 0.05).
# ----------------------------------------------------
alpha = 0.05

print("--- Hypothesis Test Results ---")
print(f"Calculated Test Statistic: {test_statistic:.4f}")
print(f"Calculated P value: {p_value:.4f}")
print(f"Significance Level (Alpha): {alpha}")

print("\n--- Decision ---")

if p_value < alpha:
    # Rule: If P value is less than 0.05, reject the Null Hypothesis
    print(f"Since the P value ({p_value:.4f}) is LESS THAN alpha ({alpha}), the result is STATISTICALLY SIGNIFICANT.")
    print("We **REJECT the Null Hypothesis ($H_0$)** and conclude that Ad B is truly making a difference.")
else:
    # Conclusion: The result could be due to random luck.
    print(f"Since the P value ({p_value:.4f}) is GREATER THAN alpha ({alpha}), we FAIL TO REJECT the Null Hypothesis ($H_0$).")
    print("We conclude that there is not enough evidence to say the difference is real.")
```

##Interpreting the Hypothesis Test Output

The T-test compared the average clicks of two ads. Below is the clean summary.

1. Data Summary and Hypotheses
Summary Table
| Metric                 | Value              |
|------------------------|--------------------|
| Ad A Average Clicks    | 48.96              |
| Ad B Average Clicks    | 55.22              |
| Observed Difference    | 6.26 (Ad B > Ad A) |

**Hypotheses**

Null Hypothesis (H0): No difference between Ad A and Ad B; the observed difference is due to chance.

Alternative Hypothesis (Ha): There is a real difference.

Significance Level (alpha): 0.05

**2. Evidence and Decision**
Test Output Table
| Result             | Value   | Interpretation                                           |
|--------------------|----------|----------------------------------------------------------|
| P Value            | 0.0000  | Highly unlikely under H0; strong evidence against H0.     |
| Decision Rule      | P < 0.05 | If P is less than alpha, result is statistically significant. |
| Final Conclusion   | Significant | We reject H0. Ad B is performing better than Ad A.        |

**Conclusion**

Since the P value (0.0000) is less than the significance level (0.05),
the result is statistically significant.
We reject the null hypothesis and conclude that Ad B truly outperforms Ad A.

```python

# This code block simulates a Normal Distribution (the bell curve), which is the second "all-star" distribution discussed.
# The process shown here uses simulation to create data that follows a specific probability blueprint.

# Step 1: Import the required Python libraries.
# NumPy is essential for generating random numbers based on a distribution.
# Matplotlib is included to visualize the data using a histogram, as described in Step 3.
import numpy as np
import matplotlib.pyplot as plt

# --- Configuration (Defining the distribution parameters) ---
# The Normal Distribution is fully defined by its center (mean, mu) and its spread (standard deviation, sigma).

# Set the average (mean, mu) for the simulation
average_mu = 65

# Set the spread (standard deviation, sigma)
spread_sigma = 8

# Set the total number of data points to generate (size of the simulation)
data_points_N = 50000

# --- Simulation Execution ---

# Step 2: Call the function numpy.random.normal.
# This function generates thousands of random numbers that adhere to the specified parameters (mu, sigma).
# 'loc' sets the mean (average_mu), 'scale' sets the standard deviation (spread_sigma).
simulated_data = np.random.normal(loc=average_mu,
                                   scale=spread_sigma,
                                   size=data_points_N)

# Step 3 & 4: Graph those numbers in a histogram and watch the bell curve appear.
# The histogram shows the frequency of the generated values.
plt.figure(figsize=(10, 6))

# Plot the data:
# 'bins=50' divides the data into 50 ranges.
# 'density=True' ensures the total area under the curve adds up to exactly one, following the rules of a Probability Density Function (PDF).
plt.hist(simulated_data, bins=50, density=True, color='#0077b6', edgecolor='black')

# Add descriptive plot elements for clarity.
plt.title('Simulated Normal Distribution (Bell Curve Appearance)')
plt.xlabel(f'Values (Center Point/Mean: {average_mu}, Spread/StDev: {spread_sigma})')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.4)

# Display the plot
plt.show()
```
