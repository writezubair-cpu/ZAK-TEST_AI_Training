## Create the main table in `README.md`

Now we will **add your row** to the class/company directory table.

Steps:

1. Run the next cell to **view** the current `README.md`.
2. Then, either:
   - Option A (easier): Use the Colab file browser (left sidebar), double-click `README.md`, and edit it manually.
   - Option B (programmatic): Let the Python code append a row for you.

For absolute beginners, manual editing (Option A) is perfectly fine.


```python
# Initial README for Employee Directory with 22 pre-allocated rows

title = "# Employee Directory\n"
description = (
    "This is a simple directory of employees in our class.\n"
    "Each participant will fill in exactly ONE row assigned to them (ID 1â€“22).\n\n"
)

# Table header
table_header = (
    "| ID | Name | Department | Role / Title |\n"
    "|----|------|------------|--------------|\n"
)

# Generate 22 empty rows
rows = []
for i in range(1, 23):  # 1 to 22 inclusive
    rows.append(f"| {i} |  |  |  |\n")

table_body = "".join(rows)

content = title + "\n" + description + table_header + table_body

with open("README.md", "w") as f:
    f.write(content)

print("âœ… Created README.md with 22 employee slots:\n")
!sed -n '1,80p' README.md
```

```python
# ðŸ‘¤ Your details
YOUR_ID         = 8
YOUR_NAME       = "Zubair"
YOUR_DEPARTMENT = "Plant Operation App"       # change if you want
YOUR_ROLE       = "Specialist"

new_row = f"| {YOUR_ID} | {YOUR_NAME} | {YOUR_DEPARTMENT} | {YOUR_ROLE} |\n"

# Read README.md
with open("README.md", "r") as f:
    lines = f.readlines()

updated_lines = []
for line in lines:
    # Strip just for checking, but keep original line formatting
    if line.strip().startswith(f"| {YOUR_ID} "):
        # Replace the entire row for this ID
        updated_lines.append(new_row)
    else:
        updated_lines.append(line)

# Write back to README.md
with open("README.md", "w") as f:
    f.writelines(updated_lines)

print("âœ… Updated row", YOUR_ID, "in README.md")
!sed -n '1,40p' README.md
```

## Optional: Push your changes back to GitHub

To send your changes from Colab to GitHub (your fork), you can use a **Personal Access Token (PAT)**.

Important:
- Treat your token like a password.
- Do not share your token with anyone.
- Do not save it in the notebook permanently.
- Use `getpass()` so it is hidden when you type.

If this feels too advanced, you can skip pushing from Colab and instead **download the changed files and upload them on GitHub manually**.

```python
from getpass import getpass

print("You only need this if you want to push your changes to GitHub from Colab.")
use_push = input("Do you want to push your changes to GitHub? (yes/no): ").strip().lower()

if use_push == "yes":
    token = getpass("Enter your GitHub Personal Access Token (input is hidden): ")

    # Set remote URL using the token (simple version for beginners)
    remote_url = f"https://{token}@github.com/{GITHUB_USERNAME}/employee-directory.git"
    !git remote set-url origin "$remote_url"

    print("Pushing to your fork on GitHub...")
    !git push origin main
    print("Push complete. Go to your GitHub repo to open a Pull Request.")
else:
    print("Skipping push. You can still download the files and upload them via the GitHub website.")
```
